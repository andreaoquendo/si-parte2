```python
import pandas as pd # para carregar o CSV
import numpy as np

from sklearn import tree
from sklearn.datasets import load_boston
from sklearn.svm import SVR
from sklearn.metrics import accuracy_score, recall_score, precision_score
```

```python
# Carrega os Dadps
sinais_vitais = pd.read_csv(r'sinais_vitais.txt')
sinais_vitais.set_index('id', inplace=True)
```

```python
sinais_vitais.head()
gravidade = sinais_vitais[['grav']].values
risco = sinais_vitais['risco'].values
```

```python
# k_fold com k=5
k = 5

grav_k = np.array_split(gravidade, k)
risco_k = np.array_split(risco, k)

clf = tree.DecisionTreeClassifier(criterion="gini", splitter="best")
```

```python
# partições
treino = {}
teste = {}

treino['grav_1'] = np.concatenate((grav_k[0],grav_k[1],grav_k[2],grav_k[3]))
treino['risco_1'] = np.concatenate((risco_k[0],risco_k[1],risco_k[2],risco_k[3]))
teste['grav_1'] = grav_k[4]
teste['risco_1'] = risco_k[4]

treino['grav_2'] = np.concatenate((grav_k[0],grav_k[1],grav_k[4],grav_k[3]))
treino['risco_2'] = np.concatenate((risco_k[0],risco_k[1],risco_k[4],risco_k[3]))
teste['grav_2'] = grav_k[2]
teste['risco_2'] = risco_k[2]

treino['grav_3'] = np.concatenate((grav_k[0],grav_k[1],grav_k[2],grav_k[4]))
treino['risco_3'] = np.concatenate((risco_k[0],risco_k[1],risco_k[2],risco_k[4]))
teste['grav_3'] = grav_k[3]
teste['risco_3'] = risco_k[3]

treino['grav_4'] = np.concatenate((grav_k[0],grav_k[4],grav_k[2],grav_k[3]))
treino['risco_4'] = np.concatenate((risco_k[0],risco_k[4],risco_k[2],risco_k[3]))
teste['grav_4'] = grav_k[1]
teste['risco_4'] = risco_k[1]

treino['grav_5'] = np.concatenate((grav_k[4],grav_k[1],grav_k[2],grav_k[3]))
treino['risco_5'] = np.concatenate((risco_k[4],risco_k[1],risco_k[2],risco_k[3]))
teste['grav_5'] = grav_k[0]
teste['risco_5'] = risco_k[0]
```

```python
clf = clf.fit(treino['grav_1'], treino['risco_1'])

prediction = clf.predict(teste['grav_1'])
print('accuracy:', accuracy_score(teste['risco_1'], prediction))
print('recall:', recall_score(teste['risco_1'], prediction, average='micro'))
print('precision:', precision_score(teste['risco_1'], prediction, average='micro'))

print(teste['grav_1'])
print(prediction)
```
